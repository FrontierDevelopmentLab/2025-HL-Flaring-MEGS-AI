```python
removed_names = set(f['File'].tolist())
#remove .fits extension if it exists
removed_names = {name.replace('.fits', '.npy') for name in removed_names}
```

```python
#move all files in the subdirectories of mixed_data to a new folder if its name is in the removed.csv file
import os
import shutil
def move_files_from_subdirs(base_dir, removed_file):
    removed_df = pd.read_csv(removed_file)
    removed_names = set(removed_df['File'].tolist())
    #remove .fits extension if it exists
    removed_names = {name.replace('.fits', '.npy') for name in removed_names}
    #print(removed_names)

    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file in removed_names:
                src_path = os.path.join(root, file)
                dest_path = os.path.join(base_dir, 'removed_files', file)
                os.makedirs(os.path.dirname(dest_path), exist_ok=True)
                shutil.move(src_path, dest_path)
                print(f"Moved: {src_path} to {dest_path}")
```

```python
def move_files_preserving_structure(base_dir, removed_names):
    """
    Move files while preserving the original directory structure

    Args:
        base_dir (str): Base directory to search in
        removed_names (list): List of filenames to remove
    """
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file in removed_names:
                src_path = os.path.join(root, file)

                # Get relative path from base_dir to preserve structure
                rel_path = os.path.relpath(root, base_dir)

                # Build destination path
                if rel_path == '.':
                    # File is directly in base_dir
                    dest_dir = os.path.join(base_dir, 'removed_files')
                else:
                    # File is in a subdirectory
                    dest_dir = os.path.join(base_dir, 'removed_files', rel_path)

                dest_path = os.path.join(dest_dir, file)

                # Ensure destination directory exists
                os.makedirs(dest_dir, exist_ok=True)

                # Move the file
                try:
                    shutil.move(src_path, dest_path)
                    print(f"Moved: {src_path} -> {dest_path}")
                except Exception as e:
                    print(f"Error moving {src_path}: {e}")
```
